```python
#function for log transformation for normalizing values ,highly skewed data
def log_tn(data,cols):
  for colname in cols:
    if (data[colname]==1.0).all():
      data[colname + '_log']=np.log(data[colname]+1)
    else:
      data[colname + '_log']=np.log(data[colname])
data.info()
```

```python
data['Mileage'] = data['Mileage'].astype(str) # Convert 'Mileage' column to string (only for non-null values)
data['Mileage'] = data.Mileage.str.split().str.get(0) #to get the numerical val only
data['Power'] = data['Power'].astype(str) # Convert 'power' column to string (only for non-null values)
data['Power'] = data.Power.str.split().str.get(0)
data['Engine'] = data['Engine'].astype(str) # Convert 'Engine' column to string (only for non-null values)
data['Engine'] = data.Engine.str.split().str.get(0)
data['New_Price'] = data['New_Price'].astype(str) # Convert 'new price' column to string (only for non-null values)
data['New_Price'] = data.New_Price.str.split().str.get(0)
```

# Filling Missing vals using mean and median

```python
data.loc[data["Mileage"]==0.0,'Mileage']=np.nan # replace the zero values Nan using np.nan
#data.Mileage.isnull().sum() # total no of nan values in mileage col
data['Mileage'].fillna(value=np.mean(data['Mileage']),inplace=True) # fill the Nan values with the mean value
# fillna() is used to fill nan vals, np.mean() to cal the mean ,inplace=True : modifies the Mileage column in place
```

```python
#for i in data['Seats']:
 # if i==0.0:
  #  print(i)
```

```python
# another method to reolace missing val by nan
data.Seats.isnull().sum()
data['Seats'].fillna(value=np.nan,inplace=True)
```

```python

# Filling missing  values using the median by 'Model' and 'Brand' group

data['Seats'] = data.groupby(['Model', 'Brand'])['Seats'].apply(lambda x: x.fillna(x.median())).reset_index(level=[0,1], drop=True)

data['Engine'] = data.groupby(['Brand', 'Model'])['Engine'].apply(lambda x: x.fillna(x.median())).reset_index(level=[0,1], drop=True)

data['Power'] = data.groupby(['Brand', 'Model'])['Power'].apply(lambda x: x.fillna(x.median())).reset_index(level=[0,1], drop=True)
```
